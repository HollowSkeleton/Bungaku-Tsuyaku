```python
import nltk
from nltk.probability import FreqDist
from nltk.tokenize import RegexpTokenizer
import matplotlib.pyplot as plt
```

```python
#nltk.download('gutenberg') #Up to date
#nltk.download('stopwords') #Up to date
```

```python
hamlet = nltk.corpus.gutenberg.words("shakespeare-hamlet.txt")
macbeth = nltk.corpus.gutenberg.words("shakespeare-macbeth.txt")
caesar = nltk.corpus.gutenberg.words("shakespeare-caesar.txt")
```

```python
stdStopWordList = nltk.corpus.stopwords.words("English")
```

```python
nltkHamlet = " ".join([word for word in hamlet if word not in stdStopWordList])
```

```python
nltkHamletLower = nltk.word_tokenize(nltkHamlet.lower())
```

```python
StopWordList = ", . / ? < > ; : ' "" - ) ( and And i I it It he He she She the The to To of Of you You a A in In he He is Is my My his His this This that That not Not him Him as As we We be Be but But".split()
```

```python
cleanHamlet = " ".join([word for word in hamlet if word not in StopWordList])
cleanMacbeth = " ".join([word for word in macbeth if word not in StopWordList])
cleanCaesar = " ".join([word for word in caesar if word not in StopWordList])
```

```python
cleanHamletLower = nltk.word_tokenize(cleanHamlet.lower())
cleanMacbethLower = nltk.word_tokenize(cleanMacbeth.lower())
cleanCaesarLower = nltk.word_tokenize(cleanCaesar.lower())
```

```python
type(cleanHamletLower)
```

```python
plt.figure(num=None, figsize=(50, 10), dpi=150, facecolor='lightgreen')
FreqDist(cleanHamletLower).plot(30)
plt.show()
```

```python
plt.figure(num=None, figsize=(50, 10), dpi=150, facecolor='lime')
FreqDist(nltkHamletLower).plot(30)
plt.show()
```

```python
plt.figure(num=None, figsize=(50, 10), dpi=150, facecolor='lightblue')
FreqDist(cleanMacbethLower).plot(30)
plt.show()
```

```python
plt.figure(num=None, figsize=(50, 10), dpi=150, facecolor='yellow')
FreqDist(cleanCaesarLower).plot(30)
plt.show()
```
